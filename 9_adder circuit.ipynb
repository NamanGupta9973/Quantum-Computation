```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
qr = 3
cr = 3
half_adder = QuantumCircuit(qr, cr) # creates a quantum circuit
```

```python
half_adder.ccx(0, 1, 2) # applying Toffoli gate
half_adder.cx(0, 1)

%matplotlib inline
half_adder.draw()
```

```python
half_adder.barrier()
half_adder.measure([0,1,2], [0,1,2]) # measure qubits and store outcome in classical bits
half_adder.draw()

# q0 and q1 are input qubits, q1 will give SUM after measurement
# q2 will give CARRY OUT after measurement
```

```python
# now we will construct a full adder

qr2 = 4
cr2 = 4
full_adder = QuantumCircuit(qr2, cr2) # creates a quantum circuit
```

```python
# for getting CARRY OUT
full_adder.ccx(1, 2, 3) # applying Toffoli gate
full_adder.ccx(0, 1, 3)
full_adder.ccx(0, 2, 3)

full_adder.barrier()

full_adder.cx(1, 2)
full_adder.cx(0, 2)

%matplotlib inline
full_adder.draw()
```

```python
full_adder.barrier()
full_adder.measure([0,1,2,3], [0,1,2,3]) # measure qubits and store outcome in classical bits
full_adder.draw()

# q0 is CARRY IN
# q1 and q2 are input qubits, q2 will give SUM after measurement
# q3 will give CARRY OUT after measurement
```
